```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cat1 = np.array([2, 3])
cat2 = np.array([-5,-8])

cats = {'cat1': cat1, 'cat2': cat2}

user_start = np.array([0,0])
```

```python
train_iters = 10
learning_rate = 0.1

def update(user, cat, learning_rate, direction):
    print('user:', user)
    print('cat:', cat)
    print('learning_rate:', (learning_rate))
    print('direction:', (direction))
    print('result:', (cat-user)*learning_rate*float(direction))
    
    return user+((cat-user)*learning_rate*float(direction))

def plot(user_pos):
    plt.figure(figsize=(6, 4))
    plt.scatter(cat1[0], cat1[1], color='blue', label='cat 1')
    plt.scatter(cat2[0], cat2[1], color='purple', label='cat 2')
    plt.scatter(user_pos[0], user_pos[1], color='red', label='user start')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)

    plt.legend(loc='upper left')
    plt.title('Plot of 3 2D Points')
    plt.grid(True)
    plt.show()
```

```python
training_data = np.array([['cat1',-1], ['cat2',1],['cat1',-1], ['cat2',1]])


user_pos = user_start
for data in training_data:
    user_pos = update(user_pos, cats[data[0]], learning_rate, data[1])
    plot(user_pos)

for data in training_data:
    user_pos = update(user_pos, cats[data[0]], learning_rate, data[1])
    plot(user_pos)
```
